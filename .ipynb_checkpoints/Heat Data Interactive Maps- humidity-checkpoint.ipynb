```python
import folium
from branca.element import Figure
import branca.colormap as cm
import pandas as pd
import numpy as np
```

```python
# read in data
data = pd.read_csv(r"..\data\BetaRelease\clean_heat_data_CH.csv")
```

```python
# Humidity

# Session 1: afternoon data
fig_h1=Figure(width=800,height=600)
m_h1=folium.Map(location=[35.920807, -79.051726],zoom_start = 13, tiles='cartodbdark_matter')
fig_h1.add_child(m_h1)


data_1 = data[data['Session'] ==1]

colormap = cm.LinearColormap(colors=['white','blue'],
                             index=[35,55],
                             vmin=35,vmax=55)
for lat,lon,hum in zip(data_1.Lat.values,data_1.Lng.values,data_1["RelativeHumidity(%)"].values):
    try:
        folium.CircleMarker([lat, lon],
                            popup="RelativeHumidity(%):"+ str(round(hum,2)) +"%",
                            radius=3,
                            color='b',
                            fill=True,
                            fill_opacity=0.5,
                            fill_color=colormap(hum),
                           ).add_to(m_h1)
    except:
        continue

        
m_h1
```
