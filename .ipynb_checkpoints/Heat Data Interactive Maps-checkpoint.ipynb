```python
import folium
from branca.element import Figure
import branca.colormap as cm
import pandas as pd
import numpy as np
```

```python
# read in data
data = pd.read_csv(r"..\data\BetaRelease\clean_heat_data_CH.csv")
```

```python
# Session 1: afternoon data
fig=Figure(width=800,height=600)
m=folium.Map(location=[35.920807, -79.051726],zoom_start = 13, tiles='cartodbdark_matter')
fig.add_child(m)
# folium.TileLayer('Stamen Terrain').add_to(m)
# folium.TileLayer('Stamen Toner').add_to(m)
# folium.TileLayer('Stamen Water Color').add_to(m)
# folium.TileLayer('cartodbpositron').add_to(m)
# folium.TileLayer('cartodbdark_matter').add_to(m)
# folium.LayerControl().add_to(m)

data_1 = data[data['Session'] ==1]

colormap = cm.LinearColormap(colors=['blue','yellow','Red'],
                             vmin=85,vmax=105)
for lat,lon,temp in zip(data_1.Lat.values,data_1.Lng.values,data_1["Temperature(°F)"].values):
     folium.CircleMarker([lat, lon],
                            popup="Temperature(°F):"+ str(round(temp,2)) +"°F",
                            radius=3,
                            color='b',
                            fill=True,
                            fill_opacity=0.5,
                            fill_color=colormap(temp),
                           ).add_to(m)
        
m
```

```python
# Session 2: afternoon data
fig2=Figure(width=800,height=600)
m2=folium.Map(location=[35.920807, -79.051726],zoom_start = 13, tiles='cartodbdark_matter')
fig2.add_child(m2)

data_2 = data[data['Session'] !=1]

colormap = cm.LinearColormap(colors=['blue','yellow','Red'],
#                              index=[85,105],
                             vmin=85,vmax=105)
for lat,lon,temp in zip(data_2.Lat.values,data_2.Lng.values,data_2["Temperature(°F)"].values):
     folium.CircleMarker([lat, lon],
                            popup="Temperature(°F):"+ str(round(temp,2)) +"°F",
                            radius=3,
                            color='b',
                            fill=True,
                            fill_opacity=0.5,
                            fill_color=colormap(temp),
                           ).add_to(m2)
        
m2
```
